```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar los 10 primeros registros con mayores activos
top_10_activos = df.nlargest(20, 'Activos Totales')

# Graficar
plt.figure(figsize=(10,6))
sns.barplot(x='Compañía', y='Activos Totales', data=top_10_activos)
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x si son muy largas
plt.title('Top 10 Compañías con Mayores Activos Totales')
plt.xlabel('Compañía')
plt.ylabel('Activos Totales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Año Fiscal', y='Total Ingreso Operativo', data=df, hue='Compañía')
plt.title('Evolución de Ingresos Operativos por Compañía')
plt.xlabel('Año Fiscal')
plt.ylabel('Total Ingreso Operativo')
plt.legend(title='Compañía')
plt.tight_layout()
plt.show()
```
