```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# Configura el WebDriver con ChromeDriverManager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# URL base de la página web
base_url = "https://www.contratos.gov.co/consultas/consultarArchivosPAAGeneral.do?selAnioPAA=2023&offset={}&nomEnti=&selMunicipioPAA=&selDepartamentoPAA="

# Rango para el offset
offset = 0
max_offset = 10  # Define el límite máximo para el offset (ajusta según el número de páginas)

# Lista para almacenar los datos
data = []

# Itera sobre las páginas cambiando el offset
while offset <= max_offset:
    # Construye la URL con el offset actual
    url = base_url.format(offset)

    # Abre la página web
    driver.get(url)

    # Espera para que la página cargue completamente
    time.sleep(5)

    # Obtener el HTML después de la interacción
    html = driver.page_source

    # Analizar el HTML con BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Seleccionamos todas las filas de la tabla desde el tercer elemento en adelante
    rows = soup.select('tbody tr')[2:]

    # Iteramos sobre cada fila y extraemos los datos
    for row in rows:
        columns = row.find_all('td')
        
        # Verificamos si hay suficientes columnas en la fila
        if len(columns) >= 7:
            entidad = columns[0].get_text(strip=True)
            año = columns[1].get_text(strip=True)
            fecha_publicacion = columns[2].get_text(strip=True)
            ubicacion_geografica = columns[3].get_text(strip=True)
            regional = columns[4].get_text(strip=True)
            
            # Extraemos valores de los campos de entrada si existen
            archivo = columns[5].find('input', {'name': 'ruta'}).get('value') if columns[5].find('input', {'name': 'ruta'}) else None
            historico = columns[6].find('input', {'name': 'codEnti'}).get('value') if columns[6].find('input', {'name': 'codEnti'}) else None

            # Agregar los datos extraídos a la lista de resultados
            data.append({
                'Entidad': entidad,
                'Año': año,
                'Fecha Publicación': fecha_publicacion,
                'Ubicación Geográfica': ubicacion_geografica,
                'Regional / Seccional': regional,
                'Archivo': archivo,
                'Histórico': historico
            })
    
    # Incrementa el offset para la próxima iteración
    offset += 1

# Cerrar el WebDriver
driver.quit()

# Convertir los datos a un DataFrame de pandas
df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo Excel
df.to_excel("datos_contratos.xlsx", index=False)
print("Datos guardados en 'datos_contratos.xlsx'")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# Configura el WebDriver con ChromeDriverManager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# URL base de la página web
base_url = "https://www.contratos.gov.co/consultas/consultarArchivosPAAGeneral.do?selAnioPAA=2023&offset={}&nomEnti=&selMunicipioPAA=&selDepartamentoPAA="

# Rango para el offset
offset = 0
max_offset = 261  # Define el límite máximo para el offset (ajusta según el número de páginas)

# Lista para almacenar los datos
data = []

# Itera sobre las páginas cambiando el offset
while offset <= max_offset:
    # Construye la URL con el offset actual
    url = base_url.format(offset)

    # Abre la página web
    driver.get(url)

    # Espera para que la página cargue completamente
    time.sleep(5)

    # Obtener el HTML después de la interacción
    html = driver.page_source

    # Analizar el HTML con BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Seleccionamos todas las filas de la tabla desde el tercer elemento en adelante
    rows = soup.select('tbody tr')[2:]

    # Verifica si se encontraron filas en la página
    if not rows:
        print(f"Error: No se encontraron registros en la página con offset={offset}")
    
    # Iteramos sobre cada fila y extraemos los datos
    for row in rows:
        columns = row.find_all('td')
        
        # Verificamos si hay suficientes columnas en la fila
        if len(columns) >= 7:
            entidad = columns[0].get_text(strip=True)
            año = columns[1].get_text(strip=True)
            fecha_publicacion = columns[2].get_text(strip=True)
            ubicacion_geografica = columns[3].get_text(strip=True)
            regional = columns[4].get_text(strip=True)
            
            # Extraemos valores de los campos de entrada si existen
            archivo = columns[5].find('input', {'name': 'ruta'}).get('value') if columns[5].find('input', {'name': 'ruta'}) else None
            historico = columns[6].find('input', {'name': 'codEnti'}).get('value') if columns[6].find('input', {'name': 'codEnti'}) else None

            # Agregar los datos extraídos a la lista de resultados
            data.append({
                'Entidad': entidad,
                'Año': año,
                'Fecha Publicación': fecha_publicacion,
                'Ubicación Geográfica': ubicacion_geografica,
                'Regional / Seccional': regional,
                'Archivo': archivo,
                'Histórico': historico
            })
    
    # Imprimir la iteración actual (sin mostrar los datos)
    print(f"Procesando página con offset={offset}")
    
    # Incrementa el offset para la próxima iteración
    offset += 1

# Cerrar el WebDriver
driver.quit()

# Verifica si se han extraído datos y muestra un mensaje si no hay registros
if not data:
    print("No se registraron datos en ninguna página.")

# Convertir los datos a un DataFrame de pandas
df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo Excel
df.to_excel("datos_contratos.xlsx", index=False)
print("Datos guardados en 'datos_contratos.xlsx'")
```
